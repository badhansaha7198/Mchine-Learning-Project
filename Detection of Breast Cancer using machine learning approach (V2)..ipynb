# Data pre-processing

# Manual encoding for diagnosis

```python
df.diagnosis.unique()
```

```python
# 'M' == 0 & 'B' == 1
df.diagnosis = df.diagnosis.replace(['M', 'B'], [0, 1])
df.head()
```

# Separating X, y

```python
X = df.drop(['diagnosis'], axis=1)
X.head()
```

```python
y = df['diagnosis']
y.head()
```

# Data scaling

```python
#Standardization
from sklearn.preprocessing import StandardScaler
std = StandardScaler()
X_std = std.fit_transform(X)
X_std
```

# Train_Test_Split 70:30

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_std, y, test_size=0.30, random_state=0)
```

# Appling algorithms

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
logr = LogisticRegression(random_state=0)
logr.fit(X_train, y_train)
logr.score(X_test, y_test)
```

```python
#Decision Tree
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier(criterion='entropy', random_state=0)
tree.fit(X_train, y_train)
tree.score(X_test, y_test)
```

```python
#Random Forest
from sklearn.ensemble import RandomForestClassifier
forest = RandomForestClassifier(n_estimators=10, criterion='entropy', random_state=0)
forest.fit(X_train, y_train)
forest.score(X_test, y_test)
```

```python
#KNN
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train, y_train)
knn.score(X_test, y_test)
```

```python
#SVM
from sklearn.svm import SVC
svm = SVC(gamma='auto')
svm.fit(X_train, y_train)
svm.score(X_test, y_test)
```

# Measuring the preformance of the algorithm (Random Forest)

```python
predict = forest.predict(X_test)
predict
```

# Accuracy (Random Forest)

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, predict)
```

# Using confusion matrix (Random Forest)

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, predict)
```

# Classification report (Random Forest)

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, predict))
```
